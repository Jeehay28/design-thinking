# 네이버 연관 검색어

```python
import requests
import json


# 서브 키워드 리스트
sub_list = ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅅ', 'ㅇ','ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ' ]

# 메인 키워드
main_keyword = input('키워드를 입력하세요 : ')

# 파일 열기
f = open(f'C:/K_DIGITAL/06. requests 심화/{main_keyword}.txt', 'w', encoding = 'utf-8')

for sub in sub_list :
    keyword = main_keyword + ' ' + sub
    response = requests.get(f'https://ac.search.naver.com/nx/ac?q={keyword}&con=0&frm=nv&ans=2&r_format=json&r_enc=UTF-8&r_unicode=0&t_koreng=1&run=2&rev=4&q_enc=UTF-8&st=100&_callback=_jsonp_6')
    data = response.text
    result = data.split('_jsonp_6(')[1][:-1]
    dic_data = json.loads(result)  #자료형 형태로 만들어주는 모듈(파이썬에서 사용할 수 있는 딕셔너리로 매칭)
    for data in dic_data.get('items')[0] :
        print(data[0])
        f.write(data[0] + '\n')


f.close()
```

## 웹크롤링으로 뉴스 기사의 제목과 날짜 수집

```python
from numpy import source
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import openpyxl

wb = openpyxl.Workbook()
ws = wb.active
ws.append(['제목', '링크', '날짜'])

#step 1 검색할 키워드 입력
query = input('검색 키워드 입력 : ')
time.sleep(5)

#step 2 크롬브라우저로 원하는 url 접속

url = 'https://www.naver.com/'
driver = webdriver.Chrome(executable_path=r'C:\K_DIGITAL/chromedriver.exe')
driver.get(url)
time.sleep(3)


#step 3 NAVER 검색창에 키워드 입력 후 Enter
search = driver.find_element(By.ID, 'query')
search.send_keys(query)
search.send_keys(Keys.RETURN)

#step 4 뉴스탭 클릭
driver.find_element_by_xpath('//*[@id="lnb"]/div[1]/div/ul/li[2]/a').click()
time.sleep(2)

#step 5 검색결과 페이지에서 원하는 뉴스기사 내용 수집 (최근 기사 4000개까지 수집 가능)
for t in range(400) : 
    news = driver.find_elements_by_css_selector ('.news_tit')
    date = driver.find_elements_by_css_selector ('.info_group > span')
    
    dates = []
    for i in date :
        dates.append(i.text)

    matching = [s for s in dates if '면' in s]
    
    for i in matching :
        dates.remove(i)
    
    print(len(dates))
            
    all_news = list (map (lambda x,y: [x,y], news, dates))
    for news in all_news :
        ws.append([news[0].text, news[0].get_attribute('href'), news[1]])
    driver.find_element_by_xpath('//*[@id="main_pack"]/div[2]/div/a[2]/i').click()
    time.sleep(2)

print('완료했습니당~')

# 저장
wb.save(f'{query}.xlsx')

# 끝
driver.quit()
```
